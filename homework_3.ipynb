```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('train.csv')
# 2. Вывести основную информацию о датасете
print("Info about dataset:")
print(df.info())  # типы данных, количество не-null
print("\nNumber of missing values per column:")
print(df.isnull().sum())  # число пропусков
print("\nBasic statistics:")
print(df.describe(include='all'))  # средние значения и др.

# 3. Посчитать процент выживаемости у каждого класса пассажиров (Pclass)
survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100
print("\nSurvival percentage by Pclass:")
print(survival_by_class)

# 4. Самое популярное мужское и женское имя на корабле
# Извлечь имена (обычно в колонке 'Name' формат "Фамилия, титул. Имя")
def extract_first_name(name):
    # Имя после титула, например: "Braund, Mr. Owen Harris" -> "Owen"
    import re
    match = re.search(r'Mr\. (\w+)', name)
    if match:
        return match.group(1)
    match = re.search(r'Mrs\. (\w+)', name)
    if match:
        return match.group(1)
    match = re.search(r'Miss\. (\w+)', name)
    if match:
        return match.group(1)
    # Если не найдено, вернуть None
    return None

df['FirstName'] = df['Name'].apply(extract_first_name)

most_popular_male_name = df[df['Sex'] == 'male']['FirstName'].mode()[0]
most_popular_female_name = df[df['Sex'] == 'female']['FirstName'].mode()[0]
print(f"\nMost popular male name: {most_popular_male_name}")
print(f"Most popular female name: {most_popular_female_name}")

# 5. Самое популярное мужское и женское имя в каждом классе
popular_names_by_class = df.groupby(['Pclass', 'Sex'])['FirstName'].agg(lambda x: x.mode().iloc[0])
print("\nMost popular male and female names by Pclass:")
print(popular_names_by_class)

# 6. Пассажиры старше 44 лет
older_than_44 = df[df['Age'] > 44]
print("\nPassengers older than 44:")
print(older_than_44)

# 7. Пассажиры моложе 44 лет и мужского пола
younger_males = df[(df['Age'] < 44) & (df['Sex'] == 'male')]
print("\nMale passengers younger than 44:")
print(younger_males)

# 8. Количество n-местных кабин (в которых было 2, 3, 4, ... человека)
# В колонке 'Cabin' могут быть несколько номеров через пробел, считаем количество человек по количеству кабин
def cabin_count(cabin):
    if pd.isna(cabin):
        return 0
    return len(cabin.split())

df['CabinCount'] = df['Cabin'].apply(cabin_count)
cabin_counts = df['CabinCount'].value_counts().sort_index()
print("\nNumber of cabins by number of occupants:")
print(cabin_counts)
```
